```python
import numpy as np
import pandas as pd
import seaborn as sns
```

#### Считаем данные из файла и поместим их в дата фрейм с помощью библиотеки pandas

```python
dt = np.dtype('f8, f8, f8, f8, U30')
iris = np.genfromtxt('iris.data', delimiter=',', dtype=dt)
iris = pd.DataFrame(iris)
iris.columns = ['SepalLength', 'SepalWidth' ,'PetalLength', 'PetalWidth', 'Species']
iris.head()
```

#### Метод describe позволяет посмотреть по каждой характеристике: 
- Количество записей
- Максимальное и минимальное значение 
- Выборочное среднее и среднеквадратическое отклонение ($\overline{X}, \delta$)
- Квартили 

```python
iris.describe()
```

#### Воспользуемся библиотекой seaborn для отображение графика в плоскости (SepalLength, SepalWidth)

```python
sns.scatterplot(data=iris, x='SepalLength', y='SepalWidth', hue='Species', style='Species')
```

#### Для отображения всех проекций данных, сгруппированных по виду воспользуемся методом pairplot

```python
sns.pairplot(iris, hue='Species', diag_kind='hist')
```
